```python
import numpy as np
import pandas as pd
```

```python
import tensorflow
from tensorflow import keras
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense,Flatten
```

```python
(X_train,y_train), (X_test,y_test) = keras.datasets.mnist.load_data()
```

```python
import matplotlib.pyplot as plt
plt.imshow(X_train[9899])
```

```python
#for the weight convergences we have to scale it down in 0-1

X_train = X_train/255
X_test = X_test/255
```

```python
X_test[0]
```

```python
#ANN

model = Sequential()

#flatten is for nd -> 1d data
model.add(Flatten(input_shape = (28,28)))
#layer_adding
model.add(Dense(128,activation = 'relu'))
model.add(Dense(32,activation = 'relu'))
model.add(Dense(10,activation = 'softmax'))
```

```python
model.summary()
```

```python
#OHE is not needed in this sparse categorical
#OHE is needed in the normal cross entropy

model.compile(loss = 'sparse_categorical_crossentropy',optimizer = 'adam', metrics=['accuracy'])

```

```python
history = model.fit(X_train,y_train,epochs = 25,validation_split = 0.2)
```

```python
y_prob = model.predict(X_test)
```

```python
l = []
for i in y_prob:
  l.append(np.argmax(i))
print(np.array(l))
```

```python
y_pred = np.array(l)
```

```python
y_pred
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test,y_pred)
```

```python
import matplotlib.pyplot as plt
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.legend(['loss','val_loss'])
plt.show()
```
