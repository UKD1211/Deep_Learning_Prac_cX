```python
import numpy as np
def NN_architecture(l:list):
  L = []
  new_l = []
  l_wb = []
  wb = []
  LLL= []
  ini_weights_biases = []
  for i in range(len(l)):
    L.append([l[i-1],l[i]])
  # print(L[1:])
  for i in L[1:]:
    new_l.append([(i[0]*i[1]),i[1]])
  print(new_l)
  arr = np.array(new_l)
  for i in range(1,arr.shape[0]+1):
    l_wb.append([f"w_{i}",f"b_{i}"])
  for i in l_wb:
    for j in i:
        wb.append(j)
  for i in new_l:
    for j in i:
        LLL.append(j)
  for i in range(len(LLL)):
    if i%2 == 0:
        ini_weights_biases.append(np.ones(LLL[i]))
    else:
        ini_weights_biases.append(np.zeros(LLL[i]))
  d = dict(zip(wb,ini_weights_biases))
  return d,L[1:]
```
