```python
import time
def fib(n):
  if n == 1 or n == 0:
    return 1
  else:
    return fib(n-1) + fib(n-2)
```

```python
start = time.time()
print(fib(39))
end = time.time()
print(end - start)
```

```python

d = {0:1, 1:1}
def mem_fib(n,d):
  if n in d:
    return d[n]
  else:
    d[n] =  mem_fib(n-1,d) + mem_fib(n-2,d)
    return d[n]
```

```python
start = time.time()
print(mem_fib(39,d))
end = time.time()
print(end - start)
```
